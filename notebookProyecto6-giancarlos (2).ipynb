```python
#5.1

# Hipotesis Xbox One vs PC
xbox_scores = df_filtered[df_filtered['platform'] == 'xone']['user_score'].dropna()
pc_scores = df_filtered[df_filtered['platform'] == 'pc']['user_score'].dropna()

alpha = 0.05

t_stat1, p_value1 = st.ttest_ind(xbox_scores, pc_scores, equal_var=False)
print("Hipótesis Xbox One vs PC")
print(f"t-statistic: {t_stat1:.4f}, p-value: {p_value1:.4f}")
if p_value1 < alpha:
    print("Conclusión: Rechazamos Hipotesis Nula → Las calificaciones promedio son diferentes.\n")
else:
    print("Conclusión: No rechazamos Hipotesis Nula → Las calificaciones promedio son iguales.\n")


```

```python
#5.2
# Hipotesis Acción vs Deportes
action_scores = df_filtered[df_filtered['genre'] == 'action']['user_score'].dropna()
sports_scores = df_filtered[df_filtered['genre'] == 'sports']['user_score'].dropna()

alpha = 0.05

t_stat2, p_value2 = st.ttest_ind(action_scores, sports_scores, equal_var=False)
print("Hipotesis Acción vs Deportes")
print(f"t-statistic: {t_stat2:.4f}, p-value: {p_value2:.4f}")
if p_value2 < alpha:
    print("Conclusión: Rechazamos Hipotesis Nula → Las calificaciones promedio son diferentes.\n")
else:
    print("Conclusión: No rechazamos Hipotesis Nula → Las calificaciones promedio son iguales.\n")

```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor            </b> <a class="tocSkip"></a>

Bien hecho con los tests, fueron realizados de forma correcta. 

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es *estadísticamente diferente*. En futuros notebooks podrías utilizar este test para decidir sobre la igualdad de varianzas. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).
